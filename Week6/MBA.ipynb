# Market-Basket Analysis

## Tạo dữ liệu dạng bảng nhị phân

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./groceries/groceries - groceries.csv')
df.fillna(0,inplace=True)
```

```python
df.head()
```

```python
products = set()

for row in df.itertuples():
    for product in row[1:]:
        products.add(product)
        
products.discard(0)
products = list(sorted(products))
```

```python
len(products)
```

```python
trans_vectors = [products]

for row in df.itertuples():
    tran_vec = []
    for product in products:
        if product in row:
            tran_vec.append(1)
        else:
            tran_vec.append(0)
    trans_vectors.append(tran_vec)

trans_vectors = np.asarray(trans_vectors)
```

```python
df = pd.DataFrame(trans_vectors)
```

```python
df.to_csv('./datasets.csv', header=None, index=None)
```
